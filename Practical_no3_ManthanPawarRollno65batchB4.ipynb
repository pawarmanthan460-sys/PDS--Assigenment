1. What is data wrangling, and why is it important before analysis?

**Data wrangling**, also known as data munging or data preparation, is the process of cleaning, structuring, and enriching raw data into a more suitable format for analysis. It involves several steps, including data collection, data exploration, data cleaning, data transformation, and data validation.

**Importance before analysis:**

*   **Ensures Data Quality:** Raw data often contains errors, inconsistencies, and missing values. Wrangling helps to identify and correct these issues, leading to more accurate and reliable analysis results.
*   **Improves Data Usability:** By structuring and transforming data, it becomes easier to work with and understand, allowing analysts to apply various analytical techniques and models effectively.
*   **Enhances Analysis Accuracy:** High-quality, well-prepared data reduces the risk of biases and misleading conclusions, leading to more robust and trustworthy insights.
*   **Saves Time and Effort:** While wrangling can be time-consuming initially, it prevents significant delays and errors during the analysis phase, ultimately saving time and effort.
*   **Facilitates Data Integration:** When combining data from multiple sources, wrangling helps to standardize formats, units, and definitions, enabling seamless integration and comprehensive analysis.
*   **Supports Machine Learning Models:** Machine learning algorithms are highly sensitive to data quality. Wrangling is crucial for creating features, handling outliers, and ensuring that the data is in a format suitable for model training and evaluation.

### 2. You receive messy data with missing values, duplicates, and wrong types. What steps will you follow to clean it?

When faced with messy data containing missing values, duplicates, and wrong data types, I would follow these systematic steps to clean it:

1.  **Understand the Data:**
    *   **Load the Data:** Load the data into a suitable structure (e.g., pandas DataFrame).
    *   **Initial Inspection:** Use `df.head()`, `df.tail()`, `df.sample()` to get a quick overview.
    *   **Check Data Types:** Use `df.info()` or `df.dtypes` to understand current data types and identify immediate mismatches.
    *   **Summary Statistics:** Use `df.describe()` for numerical columns and `df.describe(include='object')` for categorical columns to understand distributions and potential issues.

2.  **Handle Missing Values (Imputation/Removal):**
    *   **Identify Missing Values:** Use `df.isnull().sum()` to count missing values per column and `df.isnull().sum() / len(df) * 100` to get percentages.
    *   **Visualize Missing Data:** Use heatmaps (e.g., `seaborn.heatmap(df.isnull())`) to visualize patterns of missingness.
    *   **Strategy for Numerical Columns:**
        *   **Imputation:** Fill with mean, median (less sensitive to outliers), or mode. This depends on the distribution and domain knowledge.
        *   **Prediction:** Use machine learning models (e.g., K-Nearest Neighbors Imputer) for more sophisticated imputation.
    *   **Strategy for Categorical Columns:**
        *   **Imputation:** Fill with mode, a 'Missing' category, or based on other column values.
    *   **Removal:** If a column has a very high percentage of missing values (e.g., >70-80%), or if specific rows have too many missing values and are not salvageable, consider dropping those columns or rows, respectively.
    *   **Forward/Backward Fill:** For time-series data, use `ffill()` or `bfill()`.

3.  **Correct Data Types:**
    *   **Convert Types:** Use `df['column'].astype(new_type)` or `pd.to_datetime()`, `pd.to_numeric()`, `pd.Categorical()`.
    *   **Handle Errors during Conversion:** Use `errors='coerce'` in `pd.to_numeric` or `pd.to_datetime` to convert problematic values into `NaN`, which can then be handled as missing values.

4.  **Remove Duplicate Records:**
    *   **Identify Duplicates:** Use `df.duplicated().sum()` to count full-row duplicates.
    *   **Remove Duplicates:** Use `df.drop_duplicates(inplace=True)`. Consider `subset` parameter if duplicates should be identified based on a specific set of columns.

5.  **Address Inconsistent Data Entry and Standardization:**
    *   **Categorical Data:** Look for variations of the same category (e.g., 'USA', 'U.S.A.', 'United States'). Standardize them to a single representation using `replace()` or mapping.
    *   **Text Data:** Convert text to lowercase, remove extra spaces, special characters, and apply stemming/lemmatization if necessary for natural language processing tasks.
    *   **Units and Formats:** Ensure consistent units (e.g., all monetary values in USD, all temperatures in Celsius) and date formats.

6.  **Handle Outliers (Optional, Context-Dependent):**
    *   **Identify Outliers:** Use visualizations (box plots, scatter plots, histograms), statistical methods (Z-score, IQR method), or domain knowledge.
    *   **Strategy:**
        *   **Removal:** Remove outliers if they are clearly data entry errors or extreme values that would skew analysis.
        *   **Transformation:** Apply log transformation or square root transformation to reduce the impact of skewness caused by outliers.
        *   **Imputation:** Treat outliers as missing values and impute them.
        *   **Capping/Flooring:** Replace extreme values with a certain percentile value.

7.  **Data Validation and Verification:**
    *   **Cross-check with Business Rules:** Ensure the cleaned data adheres to known business rules and constraints.
    *   **Spot Checks:** Manually inspect a sample of the cleaned data.
    *   **Re-run Summary Statistics:** Check `df.info()` and `df.describe()` again to confirm the cleaning steps were successful.

### 3. How can you check column names, data types, and summary statistics? What does "info()" show that "describe()" does not?

To check column names, data types, and summary statistics in pandas:

*   **Column Names:**
    *   `df.columns`: Returns an Index object containing all column names.
    *   `df.columns.tolist()`: Converts column names to a list.

*   **Data Types:**
    *   `df.dtypes`: Returns a Series with the data type of each column.
    *   `df.info()`: Provides a concise summary of the DataFrame, including data types for each column, non-null counts, and memory usage.

*   **Summary Statistics:**
    *   `df.describe()`: Generates descriptive statistics for numerical columns (count, mean, std, min, 25%, 50%, 75%, max).
    *   `df.describe(include='object')`: Generates descriptive statistics for categorical/object columns (count, unique, top, freq).
    *   `df.describe(include='all')`: Generates descriptive statistics for all columns, including both numerical and categorical, where applicable.

#### What `df.info()` shows that `df.describe()` does not:

`df.info()` and `df.describe()` serve different but complementary purposes.

`df.info()` provides a high-level overview of the DataFrame's structure and health, specifically showing:

1.  **Non-Null Counts:** For each column, `info()` tells you the number of non-null (non-missing) entries. This is extremely valuable for quickly identifying columns with missing values and understanding their extent without needing to calculate `isnull().sum()`. `describe()` does show `count` for numerical columns, which is the number of non-null values, but `info()` provides this for **all** column types.
2.  **Memory Usage:** It reports the approximate memory usage of the DataFrame. This is crucial for performance optimization, especially with large datasets, as it helps in understanding the memory footprint of different data types.
3.  **DataFrame Index Type and Entries:** It displays the type of the index (e.g., `RangeIndex`) and the number of entries, indicating the total number of rows.
4.  **Column Names and Data Types at a glance:** While `df.dtypes` gives data types, `df.info()` presents them alongside column names and non-null counts in a very readable format, making it easy to grasp the overall structure quickly.

In contrast, `df.describe()` focuses purely on **statistical summaries** of the data within the columns, giving insights into central tendency, dispersion, and shape of the distribution, but it doesn't provide information about missing value counts across all types, or memory usage.

### 4. How do you identify missing values and outliers in a dataset?

#### Identifying Missing Values:

Missing values are common in real-world datasets and need to be identified before they can be handled. Here are common methods:

1.  **Count Missing Values per Column:**
    *   `df.isnull().sum()`: This is the most common and effective way. It returns a Series showing the count of `True` values (i.e., `NaN`, `None`, `NaT`) for each column.
    *   `df.isna().sum()`: `isna()` is an alias for `isnull()`, so it works the same way.

2.  **Calculate Percentage of Missing Values:**
    *   `df.isnull().sum() / len(df) * 100`: This provides the percentage of missing values per column, which helps prioritize columns for imputation or removal.

3.  **Visualize Missing Data:**
    *   **Heatmap (`seaborn.heatmap`):** A heatmap of `df.isnull()` or `df.isna()` can visually represent the distribution of missing values across the DataFrame. Darker areas often indicate missingness. This is good for spotting patterns (e.g., missing values always occurring together).
    *   **Missingno Library:** The `missingno` library (`msno`) provides powerful visualizations like `msno.matrix()`, `msno.bar()`, `msno.heatmap()`, and `msno.dendrogram()` to understand missingness patterns and correlations.

4.  **Check for Any Missing Values:**
    *   `df.isnull().any()`: Returns a Series indicating whether each column has *any* missing values (`True` or `False`).
    *   `df.isnull().sum().sum()`: Returns the total number of missing values in the entire DataFrame.

5.  **Identify Rows with Missing Values:**
    *   `df[df.isnull().any(axis=1)]`: Filters the DataFrame to show all rows that contain at least one missing value.

#### Identifying Outliers:

Outliers are data points that significantly deviate from other observations. They can be errors or genuine extreme values. Identifying them is crucial as they can skew statistical analyses and machine learning models. Here are common methods:

1.  **Statistical Methods:**
    *   **Z-score:** For normally distributed data, a Z-score measures how many standard deviations an element is from the mean. Values with an absolute Z-score greater than 2, 2.5, or 3 (depending on the threshold chosen) are often considered outliers. Calculated as `(x - mean) / std_dev`.
        ```python
        from scipy.stats import zscore
        df['zscore'] = np.abs(zscore(df['numerical_column']))
        outliers = df[df['zscore'] > 3]
        ```
    *   **Interquartile Range (IQR):** This method is robust to non-normal distributions. It defines outliers as values falling below `Q1 - 1.5 * IQR` or above `Q3 + 1.5 * IQR`, where Q1 is the 25th percentile, Q3 is the 75th percentile, and IQR = Q3 - Q1.
        ```python
        Q1 = df['numerical_column'].quantile(0.25)
        Q3 = df['numerical_column'].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = df[(df['numerical_column'] < lower_bound) | (df['numerical_column'] > upper_bound)]
        ```

2.  **Visual Methods:**
    *   **Box Plots:** Display the distribution of data and clearly show outliers as individual points beyond the whiskers (which typically extend to 1.5 * IQR from the quartiles).
        ```python
        import seaborn as sns
        sns.boxplot(y=df['numerical_column'])
        ```
    *   **Histograms:** Can reveal extreme values in the tails of the distribution.
        ```python
        df['numerical_column'].hist()
        ```
    *   **Scatter Plots:** For bivariate analysis, scatter plots can identify outliers that deviate significantly from the main cluster of data points.
        ```python
        sns.scatterplot(x='col_x', y='col_y', data=df)
        ```

3.  **Machine Learning-Based Methods:**
    *   **Isolation Forest:** An unsupervised algorithm that isolates anomalies by randomly selecting a feature and then randomly selecting a split value between the maximum and minimum values of the selected feature. The number of splits required to isolate a sample is equivalent to the path length from the root node to the terminating node.
    *   **Local Outlier Factor (LOF):** Measures the local deviation of density of a given sample with respect to its neighbors. It considers as outliers samples that have a substantially lower density than their neighbors.
    *   **One-Class SVM:** An unsupervised algorithm for novelty detection, where it learns a decision function for a given data set to detect new or anomalous observations.

The choice of method depends on the data distribution, the definition of an outlier in the specific context, and the overall goal of the analysis.

### 5. How do you reduce memory usage in pandas DataFrames?

Reducing memory usage in pandas DataFrames is crucial, especially when working with large datasets that might exceed available RAM, leading to slower performance or even crashes. Here's a long answer detailing various techniques:

#### 1. Choose Appropriate Data Types:

This is often the most impactful way to reduce memory usage. Pandas defaults to `int64` and `float64` for numerical data, and `object` for strings, which are often overkill.

*   **Integers (`int`):**
    *   If you know the range of your integer data, use smaller integer types: `int8`, `int16`, `int32` instead of `int64`.
    *   For example, if a column only contains values between 0 and 255, `uint8` (unsigned 8-bit integer) is sufficient.
    ```python
    df['small_int_column'] = df['small_int_column'].astype('int8')
    ```

*   **Floats (`float`):**
    *   Similarly, if precision is not paramount, use `float32` instead of `float64`.
    ```python
    df['float_column'] = df['float_column'].astype('float32')
    ```

*   **Categorical Data (`object` to `category`):**
    *   This is extremely effective for columns with a limited number of unique string values (e.g., 'Gender', 'City', 'Product Type'). Converting an `object` column to `category` stores the unique values once and uses integer pointers to them, drastically reducing memory.
    ```python
    df['categorical_column'] = df['categorical_column'].astype('category')
    ```

*   **Booleans (`object` to `bool`):**
    *   Ensure columns containing only `True`/`False` values are stored as `bool` (1 byte per value) instead of `object` (which can be much larger).
    ```python
    df['boolean_column'] = df['boolean_column'].astype('bool')
    ```

*   **Dates and Times (`object` to `datetime`):**
    *   If a column stores dates or timestamps as strings, convert it to `datetime64[ns]` using `pd.to_datetime()`. This is more memory-efficient than `object` for dates and enables time-series operations.
    ```python
    df['date_column'] = pd.to_datetime(df['date_column'])
    ```

#### 2. Read Data Efficiently:

*   **Specify `dtype` during `read_csv()`:** Instead of letting pandas infer types (which can lead to `object` types or larger numerical types), provide a `dtype` dictionary when loading data.
    ```python
    dtype = {'col1': 'int16', 'col2': 'category', 'col3': 'float32'}
    df = pd.read_csv('your_file.csv', dtype=dtype)
    ```
*   **Use `chunksize`:** For very large files that don't fit into memory, read them in chunks, process each chunk, and then concatenate or aggregate the results.
    ```python
    chunks = pd.read_csv('large_file.csv', chunksize=10000)
    processed_chunks = []
    for chunk in chunks:
        # Process chunk (e.g., reduce types, filter rows)
        processed_chunks.append(chunk)
    df = pd.concat(processed_chunks)
    ```
*   **Select only necessary columns:** If you don't need all columns, use the `usecols` parameter in `read_csv()`.
    ```python
    df = pd.read_csv('your_file.csv', usecols=['col1', 'col2'])
    ```

#### 3. Optimize String Storage:

*   **Categorical Type:** As mentioned, `category` is the best for strings with low cardinality.
*   **StringDtype:** For string columns that don't benefit from `category` (e.g., high cardinality, unique identifiers), pandas 1.0+ introduced `StringDtype` (`pd.StringDtype()`). This is more memory-efficient than the generic `object` dtype, especially for short strings, as it uses an internal array of Python strings rather than individual Python string objects.
    ```python
    df['high_cardinality_string'] = df['high_cardinality_string'].astype(pd.StringDtype())
    ```

#### 4. Delete Unnecessary Objects and Free Memory:

*   **Delete variables:** Remove DataFrames or Series that are no longer needed using `del`.
    ```python
    del old_df
    ```
*   **Force garbage collection:** Occasionally, Python's garbage collector might not immediately free up memory. You can explicitly call it:
    ```python
    import gc
    gc.collect()
    ```

#### 5. Downcasting Numerical Data:

You can write a function to automatically downcast numerical columns to their smallest possible type without losing information.

```python
def reduce_mem_usage(df, verbose=True):
    numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    start_mem = df.memory_usage().sum() / 1024**2
    for col in df.columns:
        col_type = df[col].dtypes
        if col_type in numerics:
            c_min = df[col].min()
            c_max = df[col].max()
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                else:
                    df[col] = df[col].astype(np.int64)
            else:
                if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
                    df[col] = df[col].astype(np.float16)
                elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)
    end_mem = df.memory_usage().sum() / 1024**2
    if verbose: print(f'Memory usage after optimization: {end_mem:.2f} MB (decreased by {(100 * (start_mem - end_mem) / start_mem):.1f}%)')
    return df

# Example usage:
# df = reduce_mem_usage(df)
```

#### 6. Use Sparse Data Structures:

*   If your DataFrame contains many zero or `NaN` values, pandas offers `SparseDtype` for Series and `SparseDataFrame` (though `SparseDataFrame` is deprecated in favor of `SparseArray` and `SparseDtype`). Storing data sparsely can save memory by only storing the non-zero/non-NaN values.
    ```python
    df['sparse_column'] = df['sparse_column'].astype(pd.SparseDtype("int", 0)) # for integer type with 0 as fill value
    ```

#### 7. Avoid Chaining Operations (and Create Copies Only When Necessary):

*   Chaining operations can sometimes create intermediate copies of DataFrames, consuming extra memory. Perform operations in-place or assign results to the same variable to avoid temporary copies where possible.
*   Be mindful of `.copy()` – only use it when you explicitly need a deep copy of a slice of the DataFrame; otherwise, view-based operations are more memory-efficient.

#### 8. Use Libraries Optimized for Large Datasets:

*   **Dask:** If a DataFrame truly becomes too large for pandas even after optimization, Dask provides DataFrame objects that mimic pandas but operate on out-of-core, parallelized data, allowing you to work with datasets larger than RAM.
*   **PySpark, Vaex:** Other libraries like Apache Spark (via PySpark) or Vaex are designed for large-scale data processing.

By systematically applying these techniques, you can significantly reduce the memory footprint of your pandas DataFrames, making your data analysis more efficient and manageable.

### 13. Compare merge v/s join operations.

````python
# DataFrames for demonstration
# df already exists
# df_extended already exists

print("Original df:")
display(df[['Student_ID', 'Name', 'Score_Math']].head())
print("Original df_extended:")
display(df_extended.head())

# --- MERGE ---
# pd.merge() is a standalone function that can merge DataFrames based on columns or indices.
# It's more flexible and is the recommended way for most merging operations.
merged_df_example = pd.merge(df, df_extended, on='Student_ID', how='inner')
print("\n--- Using pd.merge (inner join on 'Student_ID') ---")
print("Combines based on 'Student_ID' column. Handles overlapping column names by adding suffixes (e.g., '_x', '_y').")
display(merged_df_example)

# --- JOIN ---
# df.join() is a DataFrame method that joins DataFrames based on their indexes by default.
# It can join on a column by setting 'on' parameter, but the *other* DataFrame still uses its index.

# To use join, we often need to set an index first.
df_indexed = df.set_index('Student_ID')
df_extended_indexed = df_extended.set_index('Student_ID')

joined_df_example = df_indexed.join(df_extended_indexed, how='inner')
print("\n--- Using df.join (inner join on index) ---")
print("Combines based on index. If columns overlap, it will raise an error unless suffixes are specified.")
display(joined_df_example)

print("\n--- Key Differences ---")
print("1. **Function vs. Method**: `pd.merge()` is a function, `df.join()` is a method.")
print("2. **Default Join Key**: `merge` defaults to common columns, `join` defaults to index.")
print("3. **Flexibility**: `merge` is generally more flexible, allowing joining on multiple columns, different column names, etc.")
print("4. **Suffixes**: `merge` automatically handles overlapping column names with suffixes. `join` requires manual suffix specification or renaming for overlaps.")
````

### 14. Create custom aggregation functions.

````python
# Custom aggregation function: Range (max - min)
def score_range(series):
    return series.max() - series.min()

# Group by 'Grade' and apply custom aggregation for 'Score_Math' and built-in for 'Score_Science'
custom_agg_df = df.groupby('Grade').agg(
    Math_Score_Range=('Score_Math', score_range),
    Avg_Science_Score=('Score_Science', 'mean'),
    Count_Students=('Student_ID', 'count')
)

print("Custom aggregation by Grade:")
display(custom_agg_df)
````

15. Perform time series based data wrangling.

````python
# Create a sample time-series DataFrame
ts_data = {
    'Date': pd.to_datetime(['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-03', '2023-01-03', '2023-01-04', '2023-01-07', '2023-01-07']),
    'Sales': [100, 150, 200, 120, 180, 250, 300, 170],
    'City': ['A', 'B', 'A', 'A', 'B', 'A', 'A', 'B']
}
df_ts = pd.DataFrame(ts_data)

# 1. Set 'Date' as the index
df_ts = df_ts.set_index('Date')
print("Original Time-Series DataFrame (indexed by Date):")
display(df_ts)

# 2. Resample daily data to weekly, summing sales
# 'W' for weekly frequency, 'SM' for semi-monthly, 'M' for monthly, 'Q' for quarterly
weekly_sales = df_ts['Sales'].resample('W').sum()
print("\nWeekly Sales (summed):")
display(weekly_sales)

# 3. Calculate a 3-day rolling average of sales
# min_periods=1 allows calculation even if fewer than 3 periods are available at the start
df_ts['Sales_Rolling_Avg_3D'] = df_ts['Sales'].rolling(window='3D', min_periods=1).mean()
print("\nSales with 3-day Rolling Average:")
display(df_ts)

# 4. Forward fill missing dates (if any, after resampling)
# Create a daily range and reindex to explicitly show missing days
daily_range = pd.date_range(start=df_ts.index.min(), end=df_ts.index.max(), freq='D')
df_ts_reindexed = df_ts.reindex(daily_range)
print("\nTime-Series reindexed to daily with NaNs:")
display(df_ts_reindexed)

df_ts_ffill = df_ts_reindexed.ffill()
print("\nTime-Series after Forward Fill:")
display(df_ts_ffill)

# 5. Extract time-based features
df['Enrollment_Month'] = df['Enrollment_Date'].dt.month
df['Enrollment_DayOfWeek'] = df['Enrollment_Date'].dt.dayofweek
print("\nDataFrame with extracted time-based features (Month, DayOfWeek):")
display(df.head())
````
