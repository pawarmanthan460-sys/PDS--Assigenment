### 1. What are sequence data types in Python?

Sequence data types in Python are ordered collections of items. Each item in a sequence is assigned an index, allowing access to individual elements or parts of the sequence. Python's built-in sequence types include:

*   **Strings (`str`)**: Immutable sequences of Unicode characters.
*   **Lists (`list`)**: Mutable sequences of items, which can be of different data types.
*   **Tuples (`tuple`)**: Immutable sequences of items, similar to lists but cannot be changed after creation.
*   **Range (`range`)**: Immutable sequences of numbers, often used for looping a specific number of times.
*   **Bytes (`bytes`)** and **Bytearrays (`bytearray`)**: Sequences of bytes.

**Indexing:**

Indexing allows you to access individual elements of a sequence using their position (index). Python uses zero-based indexing, meaning the first element is at index 0.

*   **Positive Indexing**: Starts from 0 for the first element, 1 for the second, and so on.
    ```python
    my_list = [10, 20, 30, 40]
    print(my_list[0])  # Output: 10
    print(my_list[2])  # Output: 30
    ```
*   **Negative Indexing**: Starts from -1 for the last element, -2 for the second to last, and so on.
    ```python
    my_string = "Python"
    print(my_string[-1]) # Output: 'n'
    print(my_string[-3]) # Output: 'h'
    ```

**Slicing:**

Slicing allows you to extract a sub-sequence (a portion) from a sequence. It uses the syntax `[start:stop:step]`:

*   `start`: The starting index (inclusive). If omitted, defaults to 0.
*   `stop`: The ending index (exclusive). If omitted, defaults to the end of the sequence.
*   `step`: The step or stride (how many items to skip). If omitted, defaults to 1.

**Examples across different sequence types:**

```python
my_list = [10, 20, 30, 40, 50, 60]
my_tuple = (1, 2, 3, 4, 5, 6)
my_string = "ABCDEF"

# Basic Slicing
print(my_list[1:4])    # Output: [20, 30, 40] (elements from index 1 up to, but not including, index 4)
print(my_tuple[:3])    # Output: (1, 2, 3) (elements from the beginning up to index 3)
print(my_string[3:])   # Output: 'DEF' (elements from index 3 to the end)

# Slicing with Step
print(my_list[::2])    # Output: [10, 30, 50] (every second element)
print(my_tuple[1::2])  # Output: (2, 4, 6) (every second element starting from index 1)
print(my_string[::-1]) # Output: 'FEDCBA' (reverse the string)

# Negative Slicing
print(my_list[-4:-1])  # Output: [30, 40, 50]
```

### 2. What is the difference between a list and a tuple in Python? Why would you prefer a tuple over a list in real-world applications?

The main differences between lists and tuples are mutability and usage:

| Feature    | List                                  | Tuple                                    |
| :--------- | :------------------------------------ | :--------------------------------------- |
| **Mutability** | Mutable (can be changed after creation) | Immutable (cannot be changed after creation) |
| **Syntax** | Defined using square brackets `[]`    | Defined using parentheses `()`           |
| **Operations** | Supports adding, removing, changing elements | Does not support modifying elements      |
| **Performance** | Slightly slower due to overhead for mutability | Slightly faster due to immutability      |
| **Use Case** | Collections of items that might change | Collections of related but fixed items   |

**Why prefer a tuple over a list in real-world applications?**

1.  **Data Integrity (Immutability)**: If you have a collection of items that should not change throughout the program's execution (e.g., coordinates, RGB color values, database records), using a tuple ensures data integrity. It prevents accidental modification, making your code more robust and less prone to errors.
2.  **Performance**: Tuples are generally slightly faster than lists because their size is fixed, and Python can optimize their storage and access. While the difference might be negligible for small sequences, it can add up in performance-critical applications with many small collections.
3.  **Use as Dictionary Keys**: Because tuples are immutable, they can be used as keys in dictionaries (unlike lists). This is useful when you need composite keys (e.g., a `(first_name, last_name)` tuple as a key).
4.  **Function Arguments and Return Values**: Tuples are often used to return multiple values from a function or to pass a fixed set of arguments to a function, indicating that the collection of values is meant to be treated as a single, unchangeable unit.
5.  **Readability and Intent**: Using a tuple can signal to other developers (and your future self) that the data is not intended to be modified, improving code clarity and conveying design intent.

### 3. How are Python arrays different from lists? When should you use the `array` module instead of a list?

Python's built-in `list` type is very flexible; it can store elements of different data types. The `array` module, on the other hand, provides a more memory-efficient way to store sequences of elements that are all of the same basic C type.

**Differences between `array.array` and `list`:**

| Feature         | `list`                                  | `array.array`                               |
| :-------------- | :-------------------------------------- | :------------------------------------------ |
| **Element Type** | Can store elements of different types   | Stores elements of a single, uniform C type |
| **Memory Usage** | More memory-intensive (each element is a Python object) | More memory-efficient (stores raw C values) |
| **Performance** | Generally good for general-purpose use  | Faster for numeric operations, especially with large datasets |
| **Flexibility** | Highly flexible                           | Less flexible, type-restricted              |

**When to use the `array` module instead of a list:**

You should consider using the `array` module when:

1.  **Memory Efficiency is Critical**: You are dealing with a very large sequence of numbers (e.g., millions of integers or floats) and memory consumption is a concern. An array will typically use significantly less memory than a list because it stores the raw values directly rather than Python objects.
2.  **Performance for Numeric Operations**: You need to perform many numerical operations on a large sequence of homogeneous numbers. `array.array` can be faster for these operations, especially when interfacing with low-level C libraries or systems that expect C-style arrays.
3.  **Interfacing with C/C++ Code**: When you need to pass data to or receive data from C or C++ code, `array.array` provides a more direct and efficient way to represent C-style arrays.
4.  **Homogeneous Data**: Your data consists solely of a single numeric type (e.g., all integers, all floats). If you need to mix data types, a list is the appropriate choice.

For most general-purpose programming tasks where the dataset size is not excessively large or data types are mixed, Python's built-in `list` is usually sufficient and often more convenient due to its flexibility.

### 4. Explain how dictionaries work internally in Python. Why are dictionary lookups faster compared to lists or tuples?

Python dictionaries are implemented as **hash tables** (also known as hash maps). Here's a simplified explanation of how they work internally:

1.  **Hashing**: When you add a key-value pair to a dictionary, Python first calculates a `hash value` for the key. A hash function takes the key and converts it into a fixed-size integer (the hash value). For immutable objects (like numbers, strings, and tuples), the hash value is consistent. Mutable objects (like lists) cannot be dictionary keys because their hash value could change.
2.  **Buckets/Slots**: The hash table is an underlying array (or list of lists, or similar structure) divided into "buckets" or "slots." The hash value of the key determines which bucket the key-value pair will be stored in. This spreads the data across the table.
3.  **Collision Handling**: It's possible for different keys to produce the same hash value (a "hash collision"). Dictionaries handle collisions by storing multiple key-value pairs that hash to the same bucket in a linked list or similar structure within that bucket.
4.  **Lookup**: When you try to retrieve a value using a key, Python again calculates the hash value of the key. It then goes directly to the corresponding bucket. If there are collisions in that bucket, it iterates through the items in that bucket, comparing the full key (not just the hash) until it finds a match. This comparison is done using the `__eq__` method of the key.

**Why are dictionary lookups faster compared to lists or tuples?**

Dictionary lookups are significantly faster because of their hash table implementation, which provides **average O(1) (constant time) complexity** for retrieval. Here's why:

*   **Direct Access (on average)**: When you provide a key, the hash function quickly tells Python *exactly where* to look for the item (the specific bucket). Instead of searching through every item, it jumps almost directly to the location.
*   **Comparison with Lists/Tuples**: In contrast:
    *   **Lists and Tuples**: To find an item in a list or tuple by its *value*, Python has to iterate through the elements one by one from the beginning until it finds a match. In the worst case (item not found or at the end), this takes **O(n) (linear time)**, where 'n' is the number of elements. Accessing an item by its *index* in a list or tuple is O(1) because their elements are stored contiguously in memory, and the index directly translates to a memory address.
    *   However, dictionary lookups are by *key*, not by numerical index (like list indices). The power of dictionaries is finding an element by an arbitrary key efficiently.

Even with collisions, the number of items in any single bucket is usually small, so the traversal within a bucket is very fast, maintaining near-constant time performance on average. The efficiency relies on a good hash function that distributes keys evenly.

### 5. What is the difference between sets and lists in Python? How do sets handle duplicate values, and where are sets useful in practical scenarios?

Sets and lists are both collection types in Python, but they have fundamental differences in how they store and manage elements.

| Feature    | List                                  | Set                                      |
| :--------- | :------------------------------------ | :--------------------------------------- |
| **Ordering** | Ordered (elements maintain insertion order) | Unordered (elements do not have a defined order) |
| **Mutability** | Mutable (can add, remove, change elements) | Mutable (can add, remove elements, but elements themselves must be immutable) |
| **Duplicates** | Allows duplicate elements             | Does **not** allow duplicate elements    |
| **Indexing** | Supports indexing and slicing         | Does **not** support indexing or slicing |
| **Elements** | Can contain any data type             | Elements must be immutable (e.g., numbers, strings, tuples) |
| **Syntax** | Defined using square brackets `[]`    | Defined using curly braces `{}` or `set()` |

**How sets handle duplicate values:**

Sets are designed to store only unique elements. When you try to add a duplicate value to a set, it simply ignores the new addition, and the set remains unchanged. If you convert a list with duplicates into a set, the duplicate values will be automatically removed.

```python
my_list_with_duplicates = [1, 2, 2, 3, 4, 4, 5]
my_set = set(my_list_with_duplicates)
print(my_set) # Output: {1, 2, 3, 4, 5} (order is not guaranteed)

my_set.add(3) # Trying to add a duplicate '3'
print(my_set) # Output: {1, 2, 3, 4, 5} (set remains unchanged)
my_set.add(6)
print(my_set) # Output: {1, 2, 3, 4, 5, 6}
```

**Where are sets useful in practical scenarios?**

Sets are highly useful in various situations due to their unique element property and efficient operations:

1.  **Removing Duplicates**: This is one of the most common uses. If you have a list or other collection with many duplicate values and you need a collection of only the unique items, converting it to a set is the most straightforward and efficient way.
    ```python
    numbers = [1, 5, 2, 8, 5, 1, 9, 2]
    unique_numbers = list(set(numbers))
    # unique_numbers will be [1, 2, 5, 8, 9] (order may vary)
    ```
2.  **Membership Testing**: Checking if an element is present in a set is extremely fast (average O(1) time complexity), similar to dictionary lookups (because sets are implemented using hash tables internally). This is much faster than checking membership in a large list (O(n)).
    ```python
    allowed_users = {'alice', 'bob', 'charlie'}
    user = 'bob'
    if user in allowed_users:
        print(f"{user} is an allowed user.")
    ```
3.  **Mathematical Set Operations**: Sets provide efficient methods for standard mathematical set operations:
    *   **Union (`|` or `union()`):** Elements present in either set.
    *   **Intersection (`&` or `intersection()`):** Elements common to both sets.
    *   **Difference (`-` or `difference()`):** Elements in the first set but not in the second.
    *   **Symmetric Difference (`^` or `symmetric_difference()`):** Elements in either set, but not in both.

    ```python
    set_a = {1, 2, 3, 4}
    set_b = {3, 4, 5, 6}

    print(set_a | set_b)  # Union: {1, 2, 3, 4, 5, 6}
    print(set_a & set_b)  # Intersection: {3, 4}
    print(set_a - set_b)  # Difference: {1, 2}
    print(set_a ^ set_b)  # Symmetric Difference: {1, 2, 5, 6}
    ```
4.  **Checking for Subsets/Supersets**: Efficiently determine if one set is a subset or superset of another.
    ```python
    parent_set = {1, 2, 3, 4, 5}
    child_set = {2, 3}
    print(child_set.issubset(parent_set)) # Output: True
    ```

Sets are ideal when you need to manage unique collections, perform fast membership tests, or execute set-theoretic operations on data.

### Task 6: Write a program to count vowels in a given string.

````python
def count_vowels(input_string):
    vowels = "aeiouAEIOU"
    vowel_count = 0
    for char in input_string:
        if char in vowels:
            vowel_count += 1
    return vowel_count

my_text = "Programming in Python is fun"
count = count_vowels(my_text)
print(f"The string '{my_text}' has {count} vowels.")
````

### Task 7: Perform insertion and deletion operations on a list of integers.

````python
my_integer_list = [10, 20, 30, 40, 50]
print(f"Original List: {my_integer_list}")

# Insertion
my_integer_list.append(60) # Add to the end
print(f"After append(60): {my_integer_list}")

my_integer_list.insert(1, 15) # Insert 15 at index 1
print(f"After insert(1, 15): {my_integer_list}")

# Deletion
my_integer_list.remove(30) # Remove by value
print(f"After remove(30): {my_integer_list}")

popped_element = my_integer_list.pop(0) # Remove by index (and return element)
print(f"After pop(0), removed element: {popped_element}, List: {my_integer_list}")

del my_integer_list[2] # Delete by index
print(f"After del my_integer_list[2]: {my_integer_list}")
````
